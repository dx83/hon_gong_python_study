```python
import pandas as pd

data = {'name': ['Jerry','Riah','Paul'],
        'algol': ["A","A+","B"],
        'basic': ["C","B","C+"],
        'c++': ["B+","C","C+"]
        }
df = pd.DataFrame(data)
df.set_index('name', inplace=True)
df
```

```python
df.to_csv("./files/df_sample.csv")
```

```python
df.to_json("./files/df_sample.json")
```

```python
df.to_excel("./files/df_sample.xlsx")
```

```python
data1 = {'name': ['Jerry','Riah','Paul'],
        'algol': ["A","A+","B"],
        'basic': ["C","B","C+"],
        'c++': ["B+","C","C+"]
        }
data2 = {'c0': [1,2,3],
         'c1': [4,5,6],
         'c2': [7,8,9],
         'c3': [10,11,12],
         'c4': [13,14,15]}

df1 = pd.DataFrame(data1)
df1.set_index('name', inplace=True)
df1
```

```python
df2 = pd.DataFrame(data2)
df2.set_index('c0', inplace=True)
df2
```

```python
df2.columns = ['씨원','씨투','씨쓰리','씨포']
df2
```

```python
writer = pd.ExcelWriter("./files/df_excelwriter.xlsx")
df1.to_excel(writer, sheet_name="sheet1")
df2.to_excel(writer, sheet_name="sheet2")
writer.close()
```
