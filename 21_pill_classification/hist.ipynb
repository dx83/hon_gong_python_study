```python
import json

with open("pill_history.json", "r", encoding="utf-8") as f:
    history = json.load(f)
```

```python
import matplotlib.pyplot as plt

plt.plot(history["accuracy"])
plt.plot(history["val_accuracy"])

plt.legend(["train", "validation"])
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.show()
```
