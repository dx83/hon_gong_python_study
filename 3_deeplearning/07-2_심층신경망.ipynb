```python
import keras
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split
x_train_scaled = x_train / 255.0
x_train_scaled = x_train_scaled.reshape(-1, 28*28)
x_train_scaled, z_scaled, y_train, z_train = train_test_split(x_train_scaled, y_train, test_size=0.2, random_state=42)
```

```python
x_train_scaled.shape, y_train.shape, z_scaled.shape, z_train.shape
```

```python
inputs = keras.layers.Input(shape=(784,))
dense1 = keras.layers.Dense(100, activation='sigmoid')
dense2 = keras.layers.Dense(10, activation='softmax')
```

```python
model = keras.Sequential([inputs, dense1, dense2])
```

```python
model.summary()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
model = Sequential([
    Input(shape=(784,)),
    Dense(100, activation='sigmoid', name='hidden'),
    Dense(10, activation='softmax', name='output')
], name='fashion_mnist_model')
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train_scaled, y_train, epochs=5)
```

```python
from sklearn.model_selection import train_test_split
x_train_scaled = x_train / 255.0
x_train_scaled, z_scaled, y_train, z_train = train_test_split(x_train_scaled, y_train, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Flatten
model = Sequential([
    Input(shape=(28,28)),
    Flatten(),
    Dense(100, activation='relu', name='hidden'),
    Dense(10, activation='softmax', name='output')
], name='fashion_mnist_model')

model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train_scaled, y_train, epochs=5)
```

```python
model.evaluate(z_scaled, z_train)
```
