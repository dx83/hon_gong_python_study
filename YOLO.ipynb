```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "True"
```

```python
from ultralytics import YOLO

model = YOLO('./pkl/yolov8s.pt')
model.names
```

```python
import cv2

image_path = './files/boat.jpg'
img = cv2.imread(image_path)
img.shape
```

```python
result = model(img)
result
```

```python
class_ids = result[0].boxes.cls.cpu().numpy().astype(int)
confidences = result[0].boxes.conf.cpu().numpy()
```

```python
for i, cls_id in enumerate(class_ids):
    print(f"{i+1}, {model.names[cls_id]}, {confidences[i]:.2f}")
```

```python
rendered_img = result[0].plot()

cv2.imshow("yolo", rendered_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
